```python
import us

fips_code = 5

print(str("%02d" % fips_code))
print(us.states.lookup(str("%02d" % fips_code)))
print(us.states.lookup(str("%02d" % fips_code)).abbr)
```

```python
# https://github.com/unitedstates/python-us
# pip install us


for index, row in data.iterrows():
    try:
        new_state = us.states.lookup(str("%02d" % row['STATE'])).abbr
        data.loc[index,'STATE'] = str(new_state)
    except:
        pass
```

```python
# https://hausetutorials.netlify.app/posts/2020-05-14-reshaping-data-in-python-pandas/
    
long_data = pd.melt(frame=new_data, 
        id_vars=["STATE"], 
        var_name="YEAR", 
        value_name="POPESTIMATE")

long_data.head(100)
```
